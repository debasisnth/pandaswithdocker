# Median salary of the developer based on their age, Country wise

```python
df.columns
```

```python
data = df.groupby(['Country']).get_group('India')[['Age', 'ConvertedComp']]
df.fillna(0)
```

```python
data.dtypes
```

```python
data.dropna(axis='index', how='any', subset=['Age', 'ConvertedComp'], inplace=True)
```

```python
data['ConvertedComp'].unique()
```

```python
age_unique = list(data['Age'].unique())
```

```python
print(age_unique)
```

```python
len(age_unique)
```

```python
#removing unrealistic ages
age_unique = age_unique[4:30]
```

```python
filt = ( data['Age'] == 27.0 )
data[filt]['ConvertedComp'].median()
```

```python
###############
```

```python
medianSalary = []
for u_a in age_unique:
    filt = ( data['Age'] == u_a )
    medianSalary.append( data[filt]['ConvertedComp'].median()   )
    
```

## Creating data frame using multiple list

```python
d = {'UniqueAge':age_unique,'MedianSalary':medianSalary}
```
