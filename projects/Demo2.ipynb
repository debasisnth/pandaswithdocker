```python
#What % of people of each country knows python?
```

```python
total_developers = df['Country'].value_counts()
```

```python
total_developers
```

```python
numbers_of_python_devs =  df.groupby(['Country'])['LanguageWorkedWith'].apply( lambda x: x.str.contains('Python').sum() )
```

```python
# concact
```

```python
python_df = pd.concat([total_developers, numbers_of_python_devs], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country':'total_developers', 'LanguageWorkedWith': 'KnowsPython'}, inplace=True )
```

```python
python_df['PctKnowsPython'] = (python_df['KnowsPython'] / python_df['total_developers']) * 100
```

```python
python_df
```

```python
python_df.sort_values(by=['total_developers', 'PctKnowsPython'], ascending=[False, False], inplace=True)
```
